```
import control.matlab as ctl
import matplotlib.pyplot as plt
import numpy as np
```

```
num = np.array([10,10])
den = np.polymul([1,0.1,0],[1,10])
G = ctl.tf(num,den)
```

```
mag, phase, omega = ctl.bode(G)
magdB = 20*np.log10(mag)
fase_graus = 180*phase/np.pi

assint_w = np.array([0.01, 0.1, 1, 10, 100])
assint_mag_dB = np.array([60, 40, 0, -20, -60])
assint_fase_graus = np.array([-90, -135, -135, -135, -180])

plt.figure()

plt.subplot(211)
plt.title('Diagrama de Bode')
plt.semilogx(omega,magdB,label='Magnitude',linewidth=3)
plt.semilogx(assint_w,assint_mag_dB,label='Assíntotas',linestyle='dashed',color='red',marker='o')
plt.xlim(0.01,100)
plt.legend(loc='best')
plt.xlabel('Frequência, Hz')
plt.ylabel('Magnitude, dB')
plt.grid(which="both")

plt.subplot(212)
plt.semilogx(omega,fase_graus,label='Fase',linewidth=3)
plt.semilogx(assint_w,assint_fase_graus,label='Assíntotas',linestyle='dashed',color='red',marker='o')
plt.xlim(0.01,100)
plt.legend(loc='best')
plt.xlabel('Frequência, Hz')
plt.ylabel('Fase, °')
plt.grid(which="both")
```
